## Approfondimento sulle stringhe

Una stringa può essere definita con singoli apici (''), doppi apici ("")  o tripli apici ('''...''' o """...""") per scrivere su più righe.

```python
s1 = "Ciao mondo"
s2 = 'Python è fantastico'
s3 = """Questa è
una stringa
su più righe"""

print(s1)
print(s2)
print(s3)
```

**Virgolette e caratteri speciali**: a volte vogliamo inserire caratteri particolari dentro una stringa. Per farlo, usiamo il backslash (\) cioè un "escape character".

```python
s1 = "Lui ha detto: \"Ciao!\""   # \" serve per scrivere il simbolo "
s2 = 'È un\'ottima giornata'     # \' serve per scrivere '
s3 = "Prima riga\nSeconda riga"  # \n va a capo
s4 = "Tab\tSpaziato"             # \t inserisce una tabulazione

print(s1)
print(s2)
print(s3)
print(s4)
```

**String Indexing**: le stringhe sono sequenze, quindi ogni carattere ha un indice.
L'indice parte da 0 per il primo carattere.
Possiamo anche usare indici negativi (-1 è l'ultimo carattere).

```python
s = "Python"

print(s[0])   # Primo carattere -> 'P'
print(s[3])   # Quarto carattere -> 'h'
print(s[-1])  # Ultimo carattere -> 'n'
```

**Slicing Strings**: possiamo estrarre porzioni di stringa (slice) usando la sintassi s[start:end:step] o sue abbreviazioni

```python
s = "Python"

print(s[0:4])   # Caratteri da 0 a 3 -> 'Pyth'
print(s[:2])    # Dall'inizio fino a indice 2 escluso -> 'Py'
print(s[2:])    # Dal carattere 2 fino alla fine -> 'thon'
print(s[::2])   # Un carattere ogni 2 -> 'Pto'
print(s[::-1])  # Stringa invertita -> 'nohtyP'
```

**Concatenation and Repetition**: possiamo "attaccare" (concatenare) stringhe con l'operatore + e ripeterle con l'operatore *.

```python
a = "Ciao"
b = "Mondo"

print(a + " " + b)   # Concatenazione -> "Ciao Mondo"
print(a * 3)         # Ripetizione -> "CiaoCiaoCiao"
```

Attenzione! L'operatore + e * sono gli stessi per fare la somma e la moltiplicazione tra numeri.

```python
a = 3
b = 4

a + b
```

Attenzione agli errori!

```python
a = 3
b = "ciao"

a + b
```

**Metodi comuni delle stringhe**: le stringhe hanno molti metodi utili.
Alcuni dei più comuni sono .lower(), .upper(), .strip(), .replace(), .split()

```python
s = "  Python è Potente!  "

print(s.lower())      # Tutto minuscolo -> "  python è potente!  "
print(s.upper())      # Tutto maiuscolo -> "  PYTHON È POTENTE!  "
print(s.strip())      # Rimuove spazi ai lati -> "Python è Potente!"
print(s.replace("Potente", "fantastico"))  # Sostituisce una parola
print(s.split())      # Divide in lista di parole -> ['Python', 'è', 'Potente!']
print("Python".startswith("Py"))  # True se inizia con "Py"
print("Python".endswith("on"))    # True se termina con "on"
```

**Formattazione delle stringhe (metodo .format)**: possiamo inserire valori dentro le stringhe con .format()

```python
nome = "Luca"
età = 25

print("Ciao, mi chiamo {} e ho {} anni".format(nome, età))  # posizione automatica
print("Ciao, mi chiamo {0} e ho {1} anni".format(nome, età))  # posizione con indici
print("Ciao, mi chiamo {nome} e ho {anni} anni".format(nome="Luca", anni=25))  # con nomi
```

**Formatted String Literals (f-strings)**: permettono di inserire variabili direttamente dentro la stringa usando la sintassi f"...{variabile}..."

```python
nome = "Anna"
età = 30

print(f"Ciao, mi chiamo {nome} e ho {età} anni")          # Inserimento variabili
print(f"Tra 5 anni avrò {età + 5} anni")                  # Possiamo anche fare operazioni
```

**Built-in String Functions**: Python fornisce anche delle funzioni built-in che funzionano sulle stringhe.

```python
s = "Python"

print(len(s))       # Lunghezza della stringa -> 6
print(min(s))       # Carattere con valore minimo (alfabetico/ASCII) -> 'P'
print(max(s))       # Carattere con valore massimo -> 'y'
print(sorted(s))    # Restituisce una lista di caratteri ordinati -> ['P', 'h', 'n', 'o', 't', 'y']
```
