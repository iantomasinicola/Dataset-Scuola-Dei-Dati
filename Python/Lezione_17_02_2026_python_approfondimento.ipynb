## 1. Liste

Una **lista** e' una sequenza ordinata e mutabile di elementi. In data analysis le usiamo
continuamente: per raccogliere valori di una colonna, nomi di variabili, risultati di calcoli, ecc.

### 1.1 Creare una lista

````python
# Valori di fatturato mensile (in migliaia di euro)
fatturato = [120, 135, 148, 110, 162, 175, 190, 155, 143, 168, 180, 195]
fatturato
````

````python
# Lista di nomi di colonne di un ipotetico dataset
colonne = ["id_cliente", "eta", "reddito", "spesa_annua", "segmento"]
colonne
````

### 1.2 Accesso agli elementi e slicing

````python
# Primo e ultimo mese
fatturato[0], fatturato[-1]
````

````python
# Fatturato del primo trimestre (indici 0, 1, 2)
primo_trimestre = fatturato[0:3]
primo_trimestre
````

````python
# Fatturato del secondo semestre (da luglio in poi, indice 6)
secondo_semestre = fatturato[6:]
secondo_semestre
````

### 1.3 Metodi principali delle liste

`append` aggiunge un elemento in fondo alla lista.

````python
# Arriva un nuovo valore: fatturato di un mese extra (es. previsione gennaio)
fatturato.append(210)
fatturato
````

`extend` aggiunge piu' elementi da un'altra lista.

````python
# Aggiungiamo le colonne di un secondo dataset
colonne_extra = ["citta", "data_registrazione"]
colonne.extend(colonne_extra)
colonne
````

`insert` inserisce un elemento in una posizione specifica.

````python
# Inseriamo una colonna 'nome' in seconda posizione
colonne.insert(1, "nome")
colonne
````

`remove` rimuove la prima occorrenza di un valore.

````python
# Decidiamo di eliminare la colonna 'segmento'
colonne.remove("segmento")
colonne
````

`pop` rimuove e restituisce un elemento (di default l'ultimo).

````python
# Rimuoviamo l'ultimo valore di fatturato (la previsione aggiunta prima)
valore_rimosso = fatturato.pop()
valore_rimosso
````

`sort` ordina la lista (modifica in-place). `sorted()` restituisce una nuova lista ordinata.

````python
# Ordiniamo il fatturato dal mese peggiore al migliore
fatturato_ordinato = sorted(fatturato)
fatturato_ordinato
````

````python
# Ordine decrescente
sorted(fatturato, reverse=True)
````

`index` restituisce la posizione di un valore. `count` conta le occorrenze.

````python
punteggi = [8, 6, 7, 8, 9, 8, 5, 7]
punteggi.index(9)  # posizione del valore 9
````

````python
punteggi.count(8)  # quante volte compare il valore 8
````

`len`, `sum`, `min`, `max` sono funzioni built-in molto utili per l'analisi.

````python
len(fatturato), sum(fatturato), min(fatturato), max(fatturato)
````

````python
# Media del fatturato
media_fatturato = sum(fatturato) / len(fatturato)
media_fatturato
````

### 1.4 List comprehension

Le **list comprehension** sono un modo compatto per creare nuove liste a partire da quelle esistenti.
Sono molto usate per trasformare o filtrare dati.

````python
# Fatturato in euro (non in migliaia)
fatturato_euro = [x * 1000 for x in fatturato]
fatturato_euro
````

````python
# Filtrare solo i mesi con fatturato sopra la media
sopra_media = [x for x in fatturato if x > media_fatturato]
sopra_media
````

## 2. Tuple

Una **tupla** e' simile a una lista, ma e' **immutabile**: una volta creata non puo' essere modificata.
Si usano per dati che non devono cambiare, come le dimensioni di un dataset o le coordinate.

### 2.1 Creare e usare le tuple

````python
# Dimensioni di un dataset: (righe, colonne)
shape = (10000, 15)
shape
````

````python
# Accesso agli elementi
n_righe = shape[0]
n_colonne = shape[1]
n_righe, n_colonne
````

L'**unpacking** permette di assegnare i valori di una tupla a variabili separate in un colpo solo.

````python
# Unpacking delle dimensioni
righe, colonne = shape
righe, colonne
````

````python
# Tupla con statistiche riassuntive
stats = (25.3, 12.1, 4.5, 89.7)  # media, std, min, max
media, deviazione, minimo, massimo = stats
media, massimo
````

Le tuple hanno solo due metodi: `count` e `index`.

## 3. Set

Un **set** e' una collezione **non ordinata** di elementi **unici**.
Sono utili per trovare valori distinti, confrontare gruppi di dati e rimuovere duplicati.

### 3.1 Creare un set

````python
# Clienti che hanno acquistato nel mese di gennaio
clienti_gennaio = {"C001", "C002", "C003", "C004", "C005"}
clienti_gennaio
````

````python
# I set rimuovono automaticamente i duplicati
risposte = ["si", "no", "si", "si", "no", "forse", "si", "forse"]
valori_unici = set(risposte)
valori_unici
````

### 3.2 Metodi principali dei set

````python
clienti_febbraio = {"C003", "C004", "C006", "C007"}
````

`union` restituisce tutti i clienti (di entrambi i mesi).

````python
tutti_i_clienti = clienti_gennaio.union(clienti_febbraio)
tutti_i_clienti
````

`intersection` restituisce i clienti presenti in entrambi i mesi.

````python
clienti_comuni = clienti_gennaio.intersection(clienti_febbraio)
clienti_comuni
````

`difference` restituisce i clienti presenti solo nel primo set.

````python
# Clienti che hanno acquistato a gennaio ma NON a febbraio
solo_gennaio = clienti_gennaio.difference(clienti_febbraio)
solo_gennaio
````

`add` e `discard` aggiungono e rimuovono singoli elementi.

````python
clienti_gennaio.add("C008")
clienti_gennaio.discard("C001")
clienti_gennaio
````

Un uso molto comune: contare i valori unici in una lista.

````python
transazioni = ["online", "negozio", "online", "online", "negozio", "telefono", "online"]
n_canali = len(set(transazioni))
n_canali
````

## 4. Dizionari

Un **dizionario** e' una collezione di coppie **chiave-valore**.
Sono fondamentali in data analysis: rappresentano record, configurazioni, mappature tra categorie e valori.

### 4.1 Creare un dizionario

````python
# Un singolo record di un dataset clienti
cliente = {
    "id": "C001",
    "nome": "Marco Rossi",
    "eta": 34,
    "reddito": 42000,
    "segmento": "premium"
}
cliente
````

### 4.2 Accesso ai valori

````python
# Accesso diretto con la chiave
cliente["nome"]
````

````python
# Con .get() possiamo specificare un valore di default se la chiave non esiste
cliente.get("email", "non disponibile")
````

### 4.3 Metodi principali dei dizionari

`keys`, `values` e `items` restituiscono chiavi, valori e coppie.

````python
list(cliente.keys())
````

````python
list(cliente.values())
````

````python
# items() restituisce coppie (chiave, valore) - utile nei cicli for
for chiave, valore in cliente.items():
    print(f"{chiave}: {valore}")
````

Aggiungere o modificare un valore.

````python
# Aggiungere una nuova chiave
cliente["email"] = "m.rossi@email.it"

# Modificare un valore esistente
cliente["eta"] = 35
cliente
````

`update` aggiorna il dizionario con le coppie di un altro dizionario.

````python
nuovi_dati = {"citta": "Milano", "attivo": True}
cliente.update(nuovi_dati)
cliente
````

`pop` rimuove una chiave e restituisce il valore corrispondente.

````python
email_rimossa = cliente.pop("email")
email_rimossa
````

### 4.4 Dizionari come mappature

Un uso classico: mappare codici a descrizioni.

````python
# Mappa dei segmenti
mappa_segmenti = {
    1: "base",
    2: "standard",
    3: "premium",
    4: "gold"
}

codice_cliente = 3
mappa_segmenti[codice_cliente]
````

## 5. Liste di dizionari (verso Pandas)

In data analysis lavoriamo con **tabelle di dati**. Prima di usare pandas, possiamo rappresentare
una tabella come una **lista di dizionari**: ogni dizionario e' una riga, ogni chiave e' una colonna.

Questa e' esattamente la struttura che pandas usa internamente quando creiamo un DataFrame
con `pd.DataFrame(lista_di_dizionari)`.

````python
# Un piccolo dataset di vendite
vendite = [
    {"id": 1, "prodotto": "Laptop",    "prezzo": 899, "quantita": 3, "canale": "online"},
    {"id": 2, "prodotto": "Mouse",     "prezzo": 25,  "quantita": 15, "canale": "negozio"},
    {"id": 3, "prodotto": "Monitor",   "prezzo": 349, "quantita": 5, "canale": "online"},
    {"id": 4, "prodotto": "Tastiera",  "prezzo": 75,  "quantita": 8, "canale": "negozio"},
    {"id": 5, "prodotto": "Webcam",    "prezzo": 59,  "quantita": 12, "canale": "online"},
]
vendite
````

Accedere a un singolo record (riga) e a un campo specifico.

````python
# Terzo record
vendite[2]
````

````python
# Prezzo del terzo record
vendite[2]["prezzo"]
````

Estrarre una "colonna" come lista.

````python
# Tutti i prezzi
prezzi = [record["prezzo"] for record in vendite]
prezzi
````

````python
# Tutti i prodotti venduti online
online = [record["prodotto"] for record in vendite if record["canale"] == "online"]
online
````

Calcolare un totale per ogni riga.

````python
# Aggiungiamo un campo calcolato: il ricavo (prezzo * quantita)
for record in vendite:
    record["ricavo"] = record["prezzo"] * record["quantita"]

vendite
````

````python
# Ricavo totale
ricavo_totale = sum(record["ricavo"] for record in vendite)
ricavo_totale
````

## 8. Funzioni

Le **funzioni** permettono di organizzare il codice in blocchi riutilizzabili.
In data analysis le usiamo per incapsulare operazioni di pulizia, calcolo e trasformazione
che applichiamo ripetutamente.

### 8.1 Funzione senza parametri

````python
def intestazione_report():
    """Restituisce l'intestazione standard per i report."""
    return "=== REPORT ANALISI DATI ==="

intestazione_report()
````

### 8.2 Funzione con un parametro

````python
def calcola_media(valori):
    """Calcola la media aritmetica di una lista di numeri."""
    return sum(valori) / len(valori)

fatturato = [120, 135, 148, 110, 162]
calcola_media(valori = fatturato)
````

### 8.3 Funzione con piu' parametri

````python
def variazione_percentuale(valore_iniziale, valore_finale):
    """Calcola la variazione percentuale tra due valori."""
    return (valore_finale - valore_iniziale) / valore_iniziale * 100

variazione_percentuale(valore_iniziale = 120, valore_finale = 148)
````

### 8.4 Funzione con parametri di default

````python
def filtra_sopra_soglia(valori, soglia=0):
    """Restituisce solo i valori superiori alla soglia indicata."""
    return [v for v in valori if v > soglia]

ricavi = [1200, -50, 800, 3500, -200, 150]

# Uso con soglia di default (0): filtra i valori positivi
filtra_sopra_soglia(valori = ricavi)
````

````python
# Uso con soglia personalizzata
filtra_sopra_soglia(valori = ricavi, soglia=500)
````

### 8.5 Funzione con piu' valori di ritorno

Una funzione puo' restituire piu' valori come tupla.
Questo e' molto comodo per restituire statistiche riassuntive.

````python
def statistiche_base(valori):
    """Restituisce media, minimo e massimo di una lista."""
    media = sum(valori) / len(valori)
    return media, min(valori), max(valori)

punteggi = [72, 85, 91, 68, 77, 95, 83]
media, minimo, massimo = statistiche_base(valori = punteggi)

f"Media: {media:.1f}, Min: {minimo}, Max: {massimo}"
````

### 8.6 Funzione che restituisce un dizionario

````python
def profilo_colonna(valori, nome_colonna):
    """Crea un profilo statistico di base per una colonna numerica."""
    n = len(valori)
    media = sum(valori) / n
    ordinati = sorted(valori)
    return {
        "colonna": nome_colonna,
        "conteggio": n,
        "media": round(media, 2),
        "minimo": min(valori),
        "massimo": max(valori)
    }

eta_clienti = [25, 34, 45, 28, 52, 38, 41, 29, 33, 47]
profilo_colonna(valori = eta_clienti, nome_colonna = "eta")
````

### 8.7 Funzione per pulire dati

````python
def pulisci_nome_colonna(nome):
    """Normalizza il nome di una colonna: minuscolo, senza spazi esterni, underscore."""
    return nome.strip().lower().replace(" ", "_")

nomi_raw = [" Eta Cliente ", "REDDITO annuo", "  Data Registrazione  "]
nomi_puliti = [pulisci_nome_colonna(nome = n) for n in nomi_raw]
nomi_puliti
````

### 8.8 Funzione che processa una lista di dizionari

````python
def aggiungi_ricavo(records):
    """Aggiunge il campo 'ricavo' a ogni record di vendita."""
    for record in records:
        record["ricavo"] = record["prezzo"] * record["quantita"]
    return records

dati = [
    {"prodotto": "Mouse", "prezzo": 25, "quantita": 10},
    {"prodotto": "Monitor", "prezzo": 349, "quantita": 3},
    {"prodotto": "Cuffie", "prezzo": 79, "quantita": 7},
]

aggiungi_ricavo(records = dati)
````

### 8.9 Funzione che restituisce una funzione (cenno)

Un pattern utile: creare funzioni che generano altre funzioni.
Qui un semplice esempio con un convertitore di valuta.

````python
def crea_convertitore(tasso):
    """Crea una funzione che converte un importo al tasso specificato."""
    def converti(importo):
        return round(importo * tasso, 2)
    return converti

euro_to_usd = crea_convertitore(tasso = 1.08)
euro_to_usd(1000)
````

## Riepilogo

| Struttura | Ordinata | Mutabile | Duplicati | Uso tipico |
|-----------|----------|----------|-----------|------------|
| Lista     | Si       | Si       | Si        | Colonne, sequenze di dati |
| Tupla     | Si       | No       | Si        | Dati immutabili, shape, coordinate |
| Set       | No       | Si       | No        | Valori unici, confronti tra gruppi |
| Dizionario| No*      | Si       | Chiavi no | Record, mappature, configurazioni |

\* I dizionari in Python 3.7+ mantengono l'ordine di inserimento.

Queste strutture, combinate con le stringhe, le conversioni e le funzioni,
costituiscono la base per lavorare con **pandas**, dove:
- una **lista di dizionari** diventa un `DataFrame`
- le **operazioni sulle liste** diventano operazioni su `Series`
- le **funzioni** diventano trasformazioni da applicare con `.apply()`
- le **f-string** servono per generare report e output leggibili
